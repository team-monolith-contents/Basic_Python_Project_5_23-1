## **조건과 반복**

등대는 바다를 항해하는 배들을 지켜주기 위해 불빛을 내는 시설입니다.<br>
<br>
등대의 불빛이 보이는 곳엔 암초가 있어서 오지 말라는 의미를 나타냅니다.

다음 코드를 실행하면 어떻게 될까요?

```python
while True:
    print("이곳은 위험하오니 돌아가세요")
```

위 코드를 실행하면 "이곳은 위험하오니 돌아가세요"가 계속 출력됩니다.<br> 
<font color='red'>(주의) 무한반복 코드를 사용했을 때 서버가 다운될 수 있습니다.</font>

while 뒤에는 조건식이 옵니다.  
while은 **~한 동안**이라는 뜻입니다.
while을 이용해서 어떤 조건이 **참인 동안** 코드가 실행되게 만들 수 있습니다.


#### **코드가 어떻게 작동하는지 이해했나요?**
```python
while True:
```
2일차에서 배운 if에서 True라고 하는 건 항상 참이라고 했던 것을 떠올려봅시다.
조건식이 이미 참이라는 뜻입니다. 즉, 무한반복 실행하게 됩니다. 

```python
while 1 == 1:
while '사람' == '사람':
while 1 > 0:
```
모두 while True와 같이 쓸 수 있습니다.

### **비교 연산자**

while 뒤에는 조건식이 옵니다.  
같다와 같지 않다 이외에도 우리는 다양한 비교를 할 수 있습니다.

수의 크기를 비교할 때 사용하는 > 부등호는 여기에서도 똑같이 사용됩니다.  
while을 이용해서 다양한 조건을 비교하여 **참인 동안** 코드가 실행되게 만들 수 있습니다.  
다음 코드를 살펴봅시다.   

````python
#예제 5-2: while과 비교 연산자

time = int(input("몇 초 남았니?"))

while time > 0:
    print("서둘러", time, "초 남았어!" )
    time = time - 1

print("이제 출발하자")
````

#### **코드가 어떻게 작동하는지 이해했나요?**

time에 몇 초 남았는지를 저장합니다.  
int()를 이용하여 자료형도 꼼꼼하게 바꾸어줍니다.

그리고 time이 0보다 큰 경우에만 재촉하는 문장을 출력합니다.  
while문이 돌아가는 동안에 time은 1씩 줄어듭니다.

그리고 time이 0이 되면 더 이상 들여쓰기한 코드가 작동하지 않고,  
"이제 출발하자"가 출력이 됩니다.

### **비교 연산자의 종류**

2일차에 배웠던 조건을 기억해봅시다. 
같다와 틀리다 외에도  여러가지 조건이 있습니다.

비교 연산자의 종류 | 설명 | 예시
---|:---:|---
a > b | a가 b 초과일 때 참 | 17 > 5
a >= b | a가 b 이상일 때 참 | 4 >= 2
a == b | a와 b가 동일할 때 참 | '3' == '3'
a != b | a와 b가 다른 경우 참 | '7' != 7
a < b | a가 b 미만일 때 참 | 3 < 4
a <= b | a가 b 이하일 때 참 | 7 <= 9

*  <=, >=, != 에서 "="는 반드시 오른쪽에 씁니다. =>, =<, =!로 작성하지 않도록 주의하세요.
*  자료형이 다르면 다른 값이 됩니다. 예를 들어, '7'(문자)과 7(숫자)는 같지 않습니다.

### **가끔은 break가 필요해**

브레이크를 밟아!  
언제 쓰는 말인가요?  
다음 코드를 실행해봅시다.

````python
#예제 5-6: while-break 기초

speed = 0

while True:
    speed += 10
    print("현재 속도는 시속", speed, "km입니다.")
    
    if speed > 100:
        print("긴급 제동을 실시합니다.")
        break
    
    if speed > 80:
        print("슬슬 조심하세요")
    
print("과속하시면 안되요.")
````

#### **코드를 살펴봅시다**

while True:  
밑으로 들여쓰기 한 코드는 항상(무한) 반복되는 코드입니다.  

speed += 10은   
speed = speed + 10과 같은 뜻입니다.

만약 speed가 80보다 크다면 "슬슬 조심하세요"  
가 출력됩니다.

만약 speed가 100보다 크다면 "긴급 제동을 실시합니다"
가 출력되고, break가 실행됩니다.

break가 실행되면 더 이상 while문의 들여쓰기한 코드가 실행되지않고,  
while문 바깥의 코드가 실행됩니다.
